```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# We are reading our data

```python
df = pd.read_csv("ramen_ratings.csv")
df = pd.DataFrame(df)
```

# First 5 rows of our data

```python
df.head(5)
```

# Brand

```python
plt.figure(figsize = (20,5))
sns.countplot(x='Brand', data=df, palette="mako_r")
plt.show()
```

# Style

```python
plt.figure(figsize = (20,5))
sns.countplot(x='Style', data=df, palette="mako_r")
plt.show()
```

# Country

```python
plt.figure(figsize = (50,5))
sns.countplot(x='Country', data=df, palette="mako_r")
plt.show()
```
